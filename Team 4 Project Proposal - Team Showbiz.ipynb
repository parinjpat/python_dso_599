#### Description of Dataset

The metadata contains information about movies released on or before July 2017. 
This dataset includes 45,466 movies (rows) and 24 attributes (columns) collected from IMDB.

1. Adult: Movie restriction level. True: The age of audiences needs to be over 17. False: no restriction on audience age.
2. Belongs_to_collection: Collection name of a movie
3. Budget: The cost of a movie production
4. Genres: Movie types 
5. Homepage: A web page includes information about a movie.
6. Id: ID number of a movie
7. Imdb_id: ID number of a movie for internet movie database
8. Original_language: The original language of a movie
9. Original Title: The original name of a movie
10. Overview: The summary of a movie
11. Popularity: The number of people watched a movie
12. Poster_path: The path of a movie poster
13. Production_Companies: The company produced a movie 
14. Production_Countries: Which country produced a movie
15. Release Date: The date when the movie came out
16. Revenue: The revenue of a movie
17. Runtime: The length of a movie.
18. Spoken_language: The language spoken by actors in a movie
19. Status: The release status of a movie
20. Tagline: A short sentence about the movie
21. Title: The name of a movie in English
22. Video: Whether a movie has a video. 
23. Vote Average: The rating of movie evaluated by audience
24. Vote Count: The times of voting
